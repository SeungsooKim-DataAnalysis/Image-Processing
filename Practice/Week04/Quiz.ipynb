# Image Processing - PSNR Calculation

## 📌 Overview

This project computes the image quality difference between two BMP images using **MSE (Mean Squared Error)** and **PSNR (Peak Signal-to-Noise Ratio)**.

Additionally, it generates a new image by increasing the brightness of the original image and compares it with the original.

---

## 📂 Project Structure

```
Project/
│
├── Input/
│   ├── orgY.bmp        # Original image
│   └── testY.bmp       # Test image
│
├── Working/
│   └── StudentID.c     # Source code
│
└── Output/
    └── StudentID.bmp   # Brightness +40 image
```

---

## ⚙️ Program Description

1. Read two input images:

   * `orgY.bmp` (original image)
   * `testY.bmp` (test image)

2. Compute:

   * MSE and PSNR between **original and test image**

3. Generate a new image:

   * Increase brightness of `orgY.bmp` by **+40**
   * Clamp pixel values to the range **[0, 255]**
   * Save as `StudentID.bmp`

4. Compute:

   * MSE and PSNR between **original and brightness-increased image**

---

## 📊 Output

The program prints:

* MSE and PSNR between `orgY.bmp` and `testY.bmp`
* MSE and PSNR between `orgY.bmp` and `StudentID.bmp`

---

## 🧠 Key Concepts

* **MSE (Mean Squared Error)**
  Measures the average squared difference between two images.

* **PSNR (Peak Signal-to-Noise Ratio)**
  Indicates image quality. Higher values mean more similarity.

---

## 🚀 How to Run

1. Place input images in the `Input` folder
2. Compile the code:

   ```
   gcc StudentID.c -o psnr
   ```
3. Run:

   ```
   psnr.exe
   ```
4. Check results:

   * Console output (MSE, PSNR)
   * Generated image in `Output/StudentID.bmp`

---

## 📌 Notes

* This program assumes 24-bit BMP images.
* Grayscale images are handled by setting R=G=B.
* Pixel values are clipped to avoid overflow.
